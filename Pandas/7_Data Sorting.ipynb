# Sorting Data

```python
import pandas as pd

dicts1 = {'first': ['Corey', 'Jane', 'John'], 
         'last': ['Schafer', 'Doe', 'Doe'], 
         'email': ['coreyschafer@email.com', 'janedoe@email.com', 'johndoe@email.com']}

# dataframe creation
df = pd.DataFrame(dicts1)
df.head()
```

## Sorting Dataframe

```python
df.sort_values(by = 'last')
```

```python
df.sort_values(by = ['last', 'email'])
```

```python
df.sort_values(by = ['last', 'email'], ascending = [True, False], inplace=True)
df
```

```python
df.sort_index(inplace=True)
df
```

```python
df['first'].sort_values()
```
